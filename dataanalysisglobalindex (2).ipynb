
# step4-filling up missing values by "not given"

```python
data=pd.DataFrame(data,columns=['Entity','Code','Year','living_planet_index_average','living_planet_index_upper','living_planet_index_lower'],index=data.index)
```

```python
v=['living_planet_index_average','living_planet_index_upper','living_planet_index_lower']
```

```python
pdata=data.pivot_table(index=['Year'],columns=['Entity'],values=v)
```

# step8- finding real situation by tagging them {"BEST","GOOD",'BAD","WORST"}

```python
q=data.living_planet_index_average.quantile([0,0.25,0.5,0.75,1])
l=['WORST','BAD','GOOD','BEST']
```

```python
s1=pd.cut(pdata.describe().living_planet_index_average.mean(),q,labels=l)
```

```python
q=data.living_planet_index_upper.quantile([0,0.25,0.5,0.75,1])
l=['WORST','BAD','GOOD','BEST']
s2=pd.cut(pdata.describe().living_planet_index_upper.mean(),q,labels=l)
```

```python
q=data.living_planet_index_lower.quantile([0,0.25,0.5,0.75,1])
l=['WORST','BAD','GOOD','BEST']
s3=pd.cut(pdata.describe().living_planet_index_lower.mean(),q,labels=l)
```

```python
type(s1)
```

```python
situation={}
for i in s1.index:
    if s1[i]==s2[i]:
        if s1[i]==s3[i]:
            situation[i]=s1[i];
        else:
            situation[i]=s1[i]+"-"+s3[i];
    elif s1[i]!=s2[i]:
        situation[i]=s1[i]+"-"+s2[i];
    else:
        situation[i]=s1[i]+"-"+s2[i]+"-"+s3[i];
```

```python
situation=pd.DataFrame(situation,index={"real_situation"})
```

```python
situation
```
